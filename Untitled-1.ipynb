```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


data = {
    'City': ['Kolkata', 'Durgapur', 'Bardhaman', 'Siliguri', 'Asansol'],
    'Population': [1.8, 1.2, 0.9, 1.5, 1.3],
    'Infra': [9, 7.5, 6, 7, 6.8],
    'Industry': [9.5, 8.7, 5.8, 7.2, 8.2],
    'Score': [95, 88, 72, 80, 85]
}

df = pd.DataFrame(data)

X = df[['Population', 'Infra', 'Industry']]
y = df['Score']


model = LinearRegression()
model.fit(X, y)


future = pd.DataFrame({
    'City': ['Kolkata', 'Durgapur', 'Bardhaman','Siliguri','Asansol'],
    'Population': [2.0, 1.6, 1.2,1.4,1.5],
    'Infra': [9.6, 8.5, 7.0,7.8,8.2],
    'Industry': [9.8, 9.0, 6.5,7.2,8.8]
})

future['Prediction'] = model.predict(future[['Population', 'Infra', 'Industry']])


print("2035 Prediction:\n")
print(future)


plt.bar(future['City'], future['Prediction'])
plt.title("City Development 2035")
plt.show()
```
